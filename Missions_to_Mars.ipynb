# Mars Hemisphere

```python
# URL of page to be scraped
url = "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"

#browser.visit to emulate a human
browser.visit(url)
time.sleep(1)
```

```python
html = browser.html
soup = bs(html, 'html.parser')
type(soup)
```

```python
 #scraping the landing page with beautifulsoup to try out the code on the landing page
results= soup.findAll('div', attrs={"class":"description"})
hemispheres=[]
for hemisphere in results:
    hemispheres.append(hemisphere.find('h3').text)
hemispheres= ', '.join(hemispheres)
hemispheres
```

```python
#Creating the code using splinter to land on the second landing page to get full h-res image and url.
#Introduce a counter with range() function to keep track of which title and image we are targeting by numerical order.
#Change browser.find_by_css
#Grab the single element with Bs sintax
#Create a dictionary container with our findings
#Create an empty list and append the dictionaries. 

results= soup.findAll('div', attrs={"class":"description"})
num_results = len(results)
print(num_results)

hemisphere_image_urls= []

for num in range(num_results):
    browser.find_by_css("a.product-item h3")[num].click()
    html = browser.html
    soup = bs(html, 'html.parser')
    names=soup.find('h2', class_='title').get_text()
    relative_image_path=soup.find_all("img", class_="wide-image")[0]['src']
    mars_img = "https://astrogeology.usgs.gov/" + relative_image_path
    mars_dictionary= {'Title':names, 'Hemisphere':mars_img}
    hemisphere_image_urls.append(mars_dictionary)
    print(names)
    print(mars_img)
    print(mars_dictionary)
    browser.back()
```

```python
#The result is a list of dictionaries
hemisphere_image_urls
```

```python
#That we could also put in a dataframe and we could pass into SQL creating a connection, but this time I'll use MongoDB
mars_df=pd.DataFrame(hemisphere_image_urls)
mars_df
```
